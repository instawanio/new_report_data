```python
import pandas as pd
import plotly.express as px
```

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/instawanio/new_report_data/0ca344c7de81e2820d028e67071e6fcf2b4343f1/Range_age.csv")
df2
```

```python
fig = px.bar(df2, x="Range", y="Number", color="Ethnicity",title="Ethnicity age distribution during 2021",
             hover_data=["Range", "Total of Ethnicity population"])
fig.update_layout(plot_bgcolor = 'white',xaxis=dict(
        title='Range',
        tickmode='array',
        tickvals=df2["Range"],
        ticktext=[str(year) for year in df2["Range"]]
    ) )
fig.show()

fig.write_html("plot.html")
```
